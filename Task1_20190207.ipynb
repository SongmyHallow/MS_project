```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some data to be fitted
def model(x, a, b, c):
    return a * np.exp(-b * x) + c

xData = np.linspace(0, 10, 100)
y = model(xData, 2.5, 1.3, 0.5)

np.random.seed(729)
yNoise = 0.2 * np.random.normal(size=xData.size)
yData = yNoise+y

plt.plot(xData, yData)
plt.legend(['test data points'])
plt.show()
```

```python
from scipy.optimize import curve_fit
popt, pcov = curve_fit(model, xData, yData)
plt.plot(xData, model(xData, *popt), 'r-', label='fit: a=%.3f, b=%.3f, c=%.3f' %tuple (popt))
plt.plot(xData, yData, label='data: a=2.5, b=1.3, c=0.5')
plt.legend()
plt.show()
```

```python
def model2(x, a, b, c):
    return a/(x**b+c)

popt, pcov = curve_fit(model2, xData, yData)
plt.plot(xData, model2(xData, *popt), 'r-', label='fit: a=%.3f, b=%.3f, c=%.3f' %tuple (popt))
plt.plot(xData, yData, label='data: a=2.5, b=1.3, c=0.5')
plt.legend()
plt.show()
```

```python
# Minimize the summed absolute errors -- L1 norm regression
from scipy.optimize import minimize

def model3(pars, x):
    x=  np.array(x)
    a, b, c = pars
    return (a*x**b+c)

def sumError(pars):
    errs = yData - model3(pars, xData)
    return np.sum(errs**2)
sol = minimize(sumError, [1,1,1])

plt.plot(xData, model3(sol.x, xData),  'r-', label='fit: a=%.3f, b=%.3f, c=%.3f' %tuple (sol.x))
plt.plot(xData, yData, label='data: a=2.5, b=1.3, c=0.5')
plt.legend()
plt.show()
```
